```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import openpyxl
from datetime import datetime
from sklearn import metrics
from datetime import date
import re
from tqdm import tqdm_notebook as tqdm

# Read the data from an Excel file
df_more_than_3 = pd.read_excel('/Users/dawidarakelyan/Desktop/Capstone final/before_metrics.xlsx')

# Convert the data to a DataFrame
df_more_than_3 = pd.DataFrame(df_more_than_3)

# Fill null values with zero
df_more_than_3.fillna(0, inplace=True)

# Get a copy of the 'user id' column from the 'df_more_than_3' dataframe
users_id = df_more_than_3['user id'].copy()

# Drop duplicate values from the 'users_id' Series
users_id = users_id.drop_duplicates()

# iterate over each user id
for user_id in tqdm(users_id):
    # filter the dataframe by the current user id
    df_user = df_more_than_3[df_more_than_3['user id'] == user_id]
    # sort the filtered dataframe by 'Cont_start' column
    df_user.sort_values(by='Cont_start', inplace=True)
    # iterate over each row of the filtered dataframe
    for i in range(len(df_user)):
        # slice the filtered dataframe up to the current row
        df_user_i = df_user.iloc[:i+1]
        # check if the sliced dataframe has more than one row
        if len(df_user_i) >= 2:
            # calculate the count of unique 'id' values minus one
            count_value = df_user_i['id'].nunique() - 1
            # calculate the sum of previous 'Dam_id_car' values
            prev_damaged_cars = df_user_i.iloc[:-1]['Dam_id_car'].sum()
            # calculate the sum of previous 'Dam_id_human' values
            prev_damaged_human = df_user_i.iloc[:-1]['Dam_id_human'].sum()
            # calculate the sum of previous 'Dam_id_object' values
            prev_damaged_object = df_user_i.iloc[:-1]['Dam_id_object'].sum()
            # calculate the sum of previous 'Total_payoff' values
            prev_payoff = df_user_i.iloc[:-1]['Total_payoff'].sum()
            # calculate the mean of previous 'Avg_payoff' values
            average_payoff = df_user_i.iloc[:-1]['Avg_payoff'].mean()
            # calculate the minimum of previous 'Min_payoff' values
            min_payoff = df_user_i.iloc[:-1]['Min_payoff'].min()
            # calculate the maximum of previous 'Max_payoff' values
            max_payoff = df_user_i.iloc[:-1]['Max_payoff'].max()
            # calculate the sum of previous 'Total_dam_items' values
            prev_tot_damaged_items = df_user_i.iloc[:-1]['Total_dam_items'].sum()
            # calculate the sum of previous 'Total_acc' values
            prev_tot_acc = df_user_i.iloc[:-1]['Total_acc'].sum()
            # calculate the sum of previous 'Payoff_amount_car' values
            prev_pay_cars = df_user_i.iloc[:-1]['Payoff_amount_car'].sum()
            # calculate the sum of previous 'Payoff_amount_human' values
            prev_pay_human = df_user_i.iloc[:-1]['Payoff_amount_human'].sum()
            # calculate the sum of previous 'Payoff_amount_object' values
            prev_pay_object = df_user_i.iloc[:-1]['Payoff_amount_object'].sum()

            df_more_than_3.loc[(df_more_than_3['user id'] == user_id) & (df_more_than_3['Cont_start'] == df_user.iloc[i]['Cont_start']), 'prev_pay_cars'] = prev_pay_cars
            df_more_than_3.loc[(df_more_than_3['user id'] == user_id) & (df_more_than_3['Cont_start'] == df_user.iloc[i]['Cont_start']), 'prev_pay_human'] = prev_pay_human
            df_more_than_3.loc[(df_more_than_3['user id'] == user_id) & (df_more_than_3['Cont_start'] == df_user.iloc[i]['Cont_start']), 'prev_pay_object'] = prev_pay_object
            df_more_than_3.loc[(df_more_than_3['user id'] == user_id) & (df_more_than_3['Cont_start'] == df_user.iloc[i]['Cont_start']), 'prev_tot_acc'] = prev_tot_acc
            df_more_than_3.loc[(df_more_than_3['user id'] == user_id) & (df_more_than_3['Cont_start'] == df_user.iloc[i]['Cont_start']), 'prev_tot_dam_items'] = prev_tot_damaged_items
            df_more_than_3.loc[(df_more_than_3['user id'] == user_id) & (df_more_than_3['Cont_start'] == df_user.iloc[i]['Cont_start']), 'prev_contracts'] = count_value
            df_more_than_3.loc[(df_more_than_3['user id'] == user_id) & (df_more_than_3['Cont_start'] == df_user.iloc[i]['Cont_start']), 'prev_damaged_cars'] = prev_damaged_cars
            df_more_than_3.loc[(df_more_than_3['user id'] == user_id) & (df_more_than_3['Cont_start'] == df_user.iloc[i]['Cont_start']), 'prev_damaged_human'] = prev_damaged_human
            df_more_than_3.loc[(df_more_than_3['user id'] == user_id) & (df_more_than_3['Cont_start'] == df_user.iloc[i]['Cont_start']), 'prev_damaged_object'] = prev_damaged_object
            df_more_than_3.loc[(df_more_than_3['user id'] == user_id) & (df_more_than_3['Cont_start'] == df_user.iloc[i]['Cont_start']), 'prev_payoff'] = prev_payoff
            df_more_than_3.loc[(df_more_than_3['user id'] == user_id) & (df_more_than_3['Cont_start'] == df_user.iloc[i]['Cont_start']), 'prev_Avg_payoff'] = average_payoff
            df_more_than_3.loc[(df_more_than_3['user id'] == user_id) & (df_more_than_3['Cont_start'] == df_user.iloc[i]['Cont_start']), 'prev_Min_payoff'] = min_payoff
            df_more_than_3.loc[(df_more_than_3['user id'] == user_id) & (df_more_than_3['Cont_start'] == df_user.iloc[i]['Cont_start']), 'prev_Max_payoff'] = max_payoff
        else:
            df_more_than_3.loc[(df_more_than_3['user id'] == user_id) & (df_more_than_3['Cont_start'] == df_user.iloc[i]['Cont_start']), 'prev_pay_cars'] = 0
            df_more_than_3.loc[(df_more_than_3['user id'] == user_id) & (df_more_than_3['Cont_start'] == df_user.iloc[i]['Cont_start']), 'prev_pay_human'] = 0
            df_more_than_3.loc[(df_more_than_3['user id'] == user_id) & (df_more_than_3['Cont_start'] == df_user.iloc[i]['Cont_start']), 'prev_pay_object'] = 0
            df_more_than_3.loc[(df_more_than_3['user id'] == user_id) & (df_more_than_3['Cont_start'] == df_user.iloc[i]['Cont_start']), 'prev_tot_acc'] = 0
            df_more_than_3.loc[(df_more_than_3['user id'] == user_id) & (df_more_than_3['Cont_start'] == df_user.iloc[i]['Cont_start']), 'prev_tot_dam_items'] = 0
            df_more_than_3.loc[(df_more_than_3['user id'] == user_id) & (df_more_than_3['Cont_start'] == df_user.iloc[i]['Cont_start']), 'prev_contracts'] = 0
            df_more_than_3.loc[(df_more_than_3['user id'] == user_id) & (df_more_than_3['Cont_start'] == df_user.iloc[i]['Cont_start']), 'prev_damaged_cars'] = 0
            df_more_than_3.loc[(df_more_than_3['user id'] == user_id) & (df_more_than_3['Cont_start'] == df_user.iloc[i]['Cont_start']), 'prev_damaged_human'] = 0 
            df_more_than_3.loc[(df_more_than_3['user id'] == user_id) & (df_more_than_3['Cont_start'] == df_user.iloc[i]['Cont_start']), 'prev_damaged_object'] = 0 
            df_more_than_3.loc[(df_more_than_3['user id'] == user_id) & (df_more_than_3['Cont_start'] == df_user.iloc[i]['Cont_start']), 'prev_payoff'] = 0  
            df_more_than_3.loc[(df_more_than_3['user id'] == user_id) & (df_more_than_3['Cont_start'] == df_user.iloc[i]['Cont_start']), 'prev_Min_payoff'] = 0
            df_more_than_3.loc[(df_more_than_3['user id'] == user_id) & (df_more_than_3['Cont_start'] == df_user.iloc[i]['Cont_start']), 'prev_Avg_payoff'] = 0
            df_more_than_3.loc[(df_more_than_3['user id'] == user_id) & (df_more_than_3['Cont_start'] == df_user.iloc[i]['Cont_start']), 'prev_Max_payoff'] = 0
            

# Looping over each user id for the prev latest acc
for user_id in tqdm(users_id):
    
    # Selecting only the rows corresponding to the current user id
    df_user = df_more_than_3[df_more_than_3['user id'] == user_id]
    
    # Sorting the rows by the 'Cont_start' column
    df_user.sort_values(by='Cont_start', inplace=True)
    
    # Looping over each row for the current user id
    for i in range(len(df_user)):
        
        # Selecting only the rows up to and including the current row
        df_user_i = df_user.iloc[:i+1]
        
        # Checking if there are at least 2 rows
        if len(df_user_i) >= 2:
            
            # Calculating the minimum of 'Last_accident_days' for all rows except the last row
            last_acc = df_user_i.iloc[:-1]['Last_accident_days'].min()
            
            # Retrieving the value of 'Last_accident_days' for the last row
            next_acc = df_user_i.iloc[-1]['Last_accident_days']
            
            # Updating the 'prev_Latest_acc' column based on the values of 'Last_accident_days'
            if last_acc == 0:
                if next_acc != 0:
                    df_more_than_3.loc[(df_more_than_3['user id'] == user_id) & (df_more_than_3['Cont_start'] == df_user.iloc[i]['Cont_start']), 'prev-Latest_acc'] = int(next_acc)
                else:
                    df_more_than_3.loc[(df_more_than_3['user id'] == user_id) & (df_more_than_3['Cont_start'] == df_user.iloc[i]['Cont_start']), 'prev_Latest_acc'] = 0
            else:
                df_more_than_3.loc[(df_more_than_3['user id'] == user_id) & (df_more_than_3['Cont_start'] == df_user.iloc[i]['Cont_start']), 'prev_Latest_acc'] = 0

                
# Looping over each user id for the prev first acc                
for user_id in tqdm(users_id):
    # Get the dataframe for the current user
    df_user = df_more_than_3[df_more_than_3['user id'] == user_id]
    # Sort the dataframe by the contract start date
    df_user.sort_values(by='Cont_start', inplace=True)
    # Iterate through each row in the dataframe
    for i in range(len(df_user)):
        # Get the dataframe for the current row and all previous rows
        df_user_i = df_user.iloc[:i+1]
        # Check if there are at least 2 rows in the dataframe
        if len(df_user_i) >= 2:
            # Get the maximum value of the 'First_accident_days' column for all previous rows
            first_acc = df_user_i.iloc[:-1]['First_accident_days'].max()
            # Get the value of the 'First_accident_days' column for the current row
            next_acc1 = df_user_i.iloc[-1]['First_accident_days']
            # Check if the maximum value of 'First_accident_days' is 0
            if first_acc == 0:
                # If the value of 'First_accident_days' for the current row is not 0, set the value of 'prev_First_acc' to that value
                if next_acc1 != 0:
                    df_more_than_3.loc[(df_more_than_3['user id'] == user_id) & (df_more_than_3['Cont_start'] == df_user.iloc[i]['Cont_start']), 'prev_First_acc'] = int(next_acc1)
                # If the value of 'First_accident_days' for the current row is 0, set the value of 'prev_First_acc' to 0
                else:
                    df_more_than_3.loc[(df_more_than_3['user id'] == user_id) & (df_more_than_3['Cont_start'] == df_user.iloc[i]['Cont_start']), 'prev_First_acc'] = 0
            # If the maximum value of 'First_accident_days' is not 0, set the value of 'prev_First_acc' to 0
            else:
                df_more_than_3.loc[(df_more_than_3['user id'] == user_id) & (df_more_than_3['Cont_start'] == df_user.iloc[i]['Cont_start']), 'prev_First_acc'] = 0

df2=df_more_than_3.copy()


# Sort the dataframe by 'user id' and 'Cont_start' columns
df2.sort_values(['user id', 'Cont_start'], inplace=True)

# Calculate the number of previous contracts for each user
df2['Previous Contracts'] = df2.groupby('user id').cumcount()

# Convert 'Cont_start' column to datetime
df2['Cont_start'] = pd.to_datetime(df2['Cont_start'])

# Define a function to calculate age factor based on previous contracts and contract days
def calculate_age_factor(row):
    if row['Previous Contracts'] <= 0:
        return 0
    else:
        age_factor = (row['Cont_days'] / (row['Previous Contracts'] * 30)) ** 0.25
        return age_factor

# Apply the function to calculate 'Contracts Age Factor' for each row
df2['Contracts Age Factor'] = df2.apply(calculate_age_factor, axis=1)

# Calculate 'Damage Frequency' for each row
df2['Damage Frequency'] = df2['prev_tot_dam_items'] / df2['prev_tot_acc']

# Calculate 'Accident Severity' for each row
df2['Accident Severity'] = (df2['prev_payoff'] / df2['prev_tot_acc']) * (1 - df2['prev_payoff'] / df2['prev_payoff'])

# Calculate 'Payoff Frequency' for each row
df2['Payoff Frequency'] = df2['prev_payoff'] / (df2['prev_tot_acc'])

# Calculate 'Average Contract Days' for each user
df2['Average Contract Days'] = df2.groupby('user id')['Cont_days'].transform('mean')

# Calculate 'Score based on Accidents and Payoff' for each row
df2['Score based on Accidents and Payoff'] = ((1 - df2['prev_tot_acc'] / df2['prev_contracts']) * 
                      (1 - df2['prev_payoff'] / df2['prev_contracts'])) ** (1 / df2['Contracts Age Factor'])

# Replace infinite values with 0 in 'Score based on Accidents and Payoff' column
df2['Score based on Accidents and Payoff'] = np.where(np.isinf(df2['Score based on Accidents and Payoff']), 0, df2['Score based on Accidents and Payoff'])

# Calculate 'Score based on Accidents, Payoff, and Damaged Cars' for each row
df2['Score based on Accidents, Payoff, and Damaged Cars'] = ((1 - df2['prev_tot_acc'] / df2['prev_contracts']) * 
                      (1 - df2['prev_payoff'] / df2['prev_contracts']) * 
                      (1 - df2['prev_damaged_cars'] / df2['prev_tot_acc'])) ** (1 / df2['Contracts Age Factor'])

# Replace infinite values with 0 in 'Score based on Accidents, Payoff, and Damaged Cars' column
df2['Score based on Accidents, Payoff, and Damaged Cars'] = np.where(np.isinf(df2['Score based on Accidents, Payoff, and Damaged Cars']), 0, df2['Score based on Accidents, Payoff, and Damaged Cars'])

                                                                                              
# Calculate score based on accidents, payoff, and contract length
df2['Score based on Accidents, Payoff, and Contract Length'] = (
    (1 - df2['prev_tot_acc'] / df2['prev_contracts']) * 
    (1 - df2['prev_payoff'] / df2['prev_contracts']) * 
    (df2['Cont_days'] / df2['Average Contract Days']) ** (1 / df2['Contracts Age Factor'])
)

# Replace infinity with 0
df2['Score based on Accidents, Payoff, and Contract Length'] = np.where(
    np.isinf(df2['Score based on Accidents, Payoff, and Contract Length']), 0, df2['Score based on Accidents, Payoff, and Contract Length']
)

# Calculate driver tenure
df2['Driver Tenure'] = df2['Cont_days'] / 365

# Calculate overall driver score
df2['Overall Driver Score'] = (
    df2['Driver Tenure'] / 10 * 
    (1 - df2['Damage Frequency']) * 
    (1 - df2['Payoff Frequency']) * 
    df2['Score based on Accidents, Payoff, and Contract Length'] * 
    (1 - df2['Damage Frequency'])
)

# Replace infinity with 0
df2['Overall Driver Score'] = np.where(np.isinf(df2['Overall Driver Score']), 0, df2['Overall Driver Score'])

# Calculate driver score1
df2['Driver Score1'] = (
    (1 - df2['prev_tot_acc'] / df2['prev_contracts']) * 
    (1 - df2['prev_payoff'] / df2['prev_contracts']) * 
    (df2['Cont_days'] / df2['Average Contract Days']) * 
    (1 - df2['prev_damaged_cars'] / df2['prev_tot_acc']) ** (1 / df2['Contracts Age Factor'])
)

# Replace infinity with 0
df2['Driver Score1'] = np.where(np.isinf(df2['Driver Score1']), 0, df2['Driver Score1'])

# Replace negative 0 with 0
df2['Driver Score1'] = df2['Driver Score1'].replace(-0.0, 0)

# Calculate decay factor1
df2['Decay Factor1'] = 0.95 ** ((pd.to_datetime('now') - pd.to_datetime(df2['Cont_end'])).dt.days)

# Calculate driver score
df2['Driver Score'] = (
    (1 - df2['Decay Factor1']) * 
    (df2.index + 1) * 
    (df2['prev_tot_acc'] + 1) * 
    (df2['prev_payoff'] * (1 / df2['Cont_days'])) * 
    (df2['Cont_end'] - df2['Cont_start']).dt.days / df2['Average Contract Days']
)

# Columns to drop
columns_to_drop = [
    'Unnamed: 0.1', 'Unnamed: 0', 'plate_x','unique_id', 'Total_acc', 'Total_dam_items',
    'Payoff_amount_car', 'Payoff_amount_human', 'Payoff_amount_object',
    'Dam_id_car', 'Dam_id_human', 'Dam_id_object', 'Total_payoff',
    'Min_payoff', 'Avg_payoff', 'Max_payoff', 'min', 'max',
    'First_accident_days', 'Last_accident_days','Cont_end','groups','model']

df2 = df2.drop(columns=columns_to_drop)
                                                                                               
df2.to_excel('/Users/dawidarakelyan/Desktop/Capstone final/testing_findal.xlsx')                                                                                               
```
